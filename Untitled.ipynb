```python
import pandas as pd
import geopandas as gpd
```

```python
df = gpd.read_file("ksh-budapest.json")
```

```python
some_data_available = df[df['data'].notna()]
```

```python
sample = {'total': {'2015': {'ar': 216, 'db': 4},
  '2016': {'ar': 302, 'db': 6},
  '2017': {'ar': 294, 'db': 8},
  '2018': {'ar': 367, 'db': 7},
  '2019': {'ar': 359, 'db': 4},
  '2020': {'ar': 434, 'db': 3},
  '2021': {'ar': 504, 'db': 5},
  '2023': {'ar': 696, 'db': 4},
  '2024': {'ar': 772, 'db': 6}}}
```

```python
pd.json_normalize(sample.get('total'))
```

```python
prices = pd.json_normalize(some_data_available['data'])
```

```python
merged = pd.concat([some_data_available, prices], axis=1)
```

```python
merged.to_csv('merged.csv', index=False)
```
