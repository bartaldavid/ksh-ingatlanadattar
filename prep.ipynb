```python
import pandas as pd
```

```python
telepulesek = pd.read_csv('teleplist.csv')
```

```python
agglo_top_10 = [
    "Érd",
    "Szentendre",
    "Dunakeszi",
    "Budaörs",
    "Gödöllő",
    "Vác",
    "Szigetszentmiklós",
    "Dunaharaszti",
    "Gyál",
    "Göd",
    'Budapest'
]
telids=telepulesek[telepulesek['nev'].isin(agglo_top_10)]['id'].tolist()
```

```python
ingatlanarak= pd.read_csv('hu-ingatlan.csv')
```

```python
merged =pd.merge(ingatlanarak, telepulesek, left_on='telaz', right_on='id', how='inner')
```

```python
merged_agglo = merged[merged['telaz'].isin(telids) & (merged['kozter'] == 'együtt')] 
```

```python
merged_agglo[['ev', 'nev', 'total_ar', 'total_db']].to_csv('agglo_top_10_total.csv', index=False)
```

```python
merged_agglo.pivot(index='ev', columns='nev', values='total_ar').to_csv('agglo_top_10_pivot.csv')
```

```python
merged_agglo.pivot(index='ev', columns='nev', values='total_db').to_csv('agglo_top_10_pivot_db.csv')
```
